```python
#Use the pandas '.read_csv()' method to assign the contents of the Spotify dataset to a variable
#called songs

import pandas as pd 

#assign the spotify path to a variable
spotify_file="./data/spotify.csv"

#assign the contents of the Spotify dataset to a variable called songs
songs=pd.read_csv(spotify_file,header=0)

#Use 'head()' to show the first five rows of 'songs'
songs.head()
```

```python
#Convert the values in the 'duration_ms' column from strings to integers
import numpy as np 
songs=pd.read_csv(spotify_file,header=0,dtype={'duration_ms':int})

print(songs.dtypes)
```

```python
#Convert the values in the 'duration_ms' column from milliseconds to minutes by 
#dividing each value by 60000

def change_duration (duration):
    converted = duration/60000
    return converted

# run the function to update the values in the duration_ms column
songs["duration_ms"]=songs["duration_ms"].map(change_duration)

songs.head()
```

```python
#use 'rename()' to change the name of the 'duration_ms' column to 'minutes_long'
songs=songs.rename(columns={'duration_ms':'minutes_long'})

songs.head()
```

```python
#Use '.count()' to show how many times 'FKA twigs' is a value in the 'artists' column
twigs_count=songs[songs.artists=='FKA twigs']
print(twigs_count.count())
```

```python
#Use 'groupby()' to get the max 'danceability' rating for each 'track_genre'ArithmeticError

max_danceability = songs.groupby('track_genre').danceability.max()

print(max_danceability)
```
